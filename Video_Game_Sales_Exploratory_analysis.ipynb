```python
#Loading data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
import io
df2 = pd.read_csv(io.BytesIO(uploaded['Video_Games_Sales_as_at_22_Dec_2016.csv']))
```

```python
#Viewing columns of data
df2.head()
#The top 5 best-selling games of all time are printed in the result. The best selling game of all-time is Wii Sports for the Nintendo Wii which released in 2006.
```

```python
#Looking at and dropping null data
df2.isnull().sum()
df2 = df2.dropna()
df2.info()
```

```python
#Count of number of games published by each publisher 
print(df2['Publisher'].value_counts())
print(df2['Publisher'].nunique())
#Electronic Arts has published the most games, followed by Activision, Namco Bandai, Ubisoft, and Konami.
```

```python
#Count the number of games developed by each publisher
print(df2['Developer'].value_counts())
print(df2['Developer'].nunique())
#Ubisoft has developed the most games of all-time at 204,
#followed by EA Sports at 172 games published, EA Canada at 167 game, Konami at 162, and Capcom at 139.
```

```python
#Trend of video game sales from 1995-2016
all_time = df2.groupby('Year_of_Release')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(17,17))
plt.bar(all_time.index,all_time)
plt.xlabel('Year')
plt.ylabel('Number of games sold in millions')
plt.title('Global Video Game Sales By Year')
plt.show()
#The graph shows a steady rise in global sales, with sales peaking from 2008-09, and steadily dropping in the years following. 
#One reason for this steep decline in sales could be explained by the fact that newer games have not had as much time to sell compared to older games.
#As a result their sales are lower.
```

```python
#Bar chart of the count of video games sold by genre all-time
gat = df2.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(13,13))
plt.bar(gat.index,gat)
plt.xlabel("Genre")
plt.ylabel("Count of games sold globally (in millions)")
plt.title("Global Sales of Video Games By Genre")
plt.show()
#Action game have sold the most number of games at around 1200 
#followed by sports at around 800, shooters at a little less than 800 
#Next, role-playing sold around 450, and racing sold a little less than 450.
```

```python
#Number of games on the best selling list by platform
import seaborn as sns
plat_count = pd.crosstab(df2.Platform,df2.Global_Sales)
plat_count_total = plat_count.sum(axis=1).sort_values(ascending=False)
plt.figure(figsize=(8,8))
sns.barplot(y = plat_count_total.index, x = plat_count_total.values, orient='h')
plt.xlabel('Global Sales in millions')
plt.ylabel('Platform')
plt.title("Global Video Game Sales by platform")
plt.show()
#The graph shows that the PS2 has sold the most games, followed by the Nintendo DS, the PS3, and the Nintendo Wii.
```

```python
#Percentage of total sales by region
data = df2
top_sale_reg = data[['NA_Sales', 'EU_Sales','JP_Sales','Other_Sales']]
top_sale_reg = top_sale_reg.sum().reset_index()
top_sale_reg = top_sale_reg.rename(columns={"index": "region", 0: "sale"})
top_sale_reg
labels = top_sale_reg['region']
sizes = top_sale_reg['sale']
plt.figure(figsize=(10,8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Share of Global Sales by Region")
#North American sales have accounted for more than half of the global sales all-time, followed by European sales, other sales, and Japanese sales.
```

```python
#Critic score vs global sales
plt.scatter(df2.Critic_Score,df2.Global_Sales)
plt.xlabel("Critic score")
plt.ylabel("Global Sales")
plt.title("Critic Score vs Global Sales")
plt.show()
#There does not seem to be a correlation between the critic score and global sales.


```

```python
#Looking at the sales of video games regionally by genre
group_genre = df2.groupby(['Genre']).sum().loc[:, 'NA_Sales': 'Global_Sales']
group_genre['NA_Sales%'] = group_genre['NA_Sales']/group_genre['Global_Sales']
group_genre['EU_Sales%'] = group_genre['EU_Sales']/group_genre['Global_Sales']
group_genre['JP_Sales%'] = group_genre['JP_Sales']/group_genre['Global_Sales']
group_genre['Other_Sales%'] = group_genre['Other_Sales']/group_genre['Global_Sales']
plt.figure(figsize=(10,10))
sns.set(font_scale=0.7)
plt.subplot(211)
sns.heatmap(group_genre.loc[:, 'NA_Sales': 'Other_Sales'], annot=True, fmt='.1f')
plt.title("Regional Sales of Video Games by Genre")
plt.subplot(212)
sns.heatmap(group_genre.loc[:, 'NA_Sales%':'Other_Sales%'], vmax=1, vmin=0, annot=True, fmt = '.2%')
plt.title("Percentage of Regional Sales of Video Games by Genre")
plt.show()
#The generated heatmap shows that North America led in the sales of Action games by a large margin. 
#The insight that can be taken from this is that Action games are extremely popular in North America, furthermore they were the highest selling genre overall.
#North America is the best-selling region for all 12 video game genres. 
#This is in line with the pie chart visualization which shows that North America is responsible for nearly half of the all-time total global sales.
#The most popular game genre in Europe and Other Regions is action games.
#The most popular game genre 
```
